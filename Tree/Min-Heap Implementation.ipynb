```python
#Min-Heap Implementation

class BinaryHeap():
    
    def __init__(self):
        self.heaplist=[0]
        self.currentsize=0
        
    def insert(self, k):
        self.heaplist.append(k)
        self.currentsize=self.currentsize+1
        self.percup(self.currentsize)
    
    #percup is to maintain the heap property affter insertion.
    def percup(self, i):
        while i//2>0:
            if self.heaplist[i//2]>self.heaplist[i]:
                temp = self.heaplist[i//2]
                self.heaplist[i//2]=self.heaplist[i]
                self.heaplist[i]=temp
                
            i=i//2
            
            
    def delmin(self):
        retval = self.heaplist[1]  #Item Deleted
        self.heaplist[1]=self.heaplist[self.currentsize]
        self.currentsize=self.currentsize-1
        self.heaplist.pop()
        self.percdown(1) #AFTER DELEion moves last element to root and then maintain min-heap property.
        return retval
    
    def percdown(self,i):
        while(i*2<=self.currentsize):
            mc=self.minchild(i)
            if self.heaplist[i] > self.heaplist[mc]:
                tmp = self.heaplist[i]
                self.heaplist[i] = self.heaplist[mc]
                self.heaplist[mc] = tmp
            i = mc


            
    def minchild(self, i):
        if i*2+1>self.currentsize:
            return i*2
        else:
            if self.heaplist[i*2]<self.heaplist[i*2+1]:
                return i*2
            else:
                return i*2+1
        
    
            
            
heap=BinaryHeap()
heap.insert(5)
heap.insert(9)
heap.insert(11)
heap.insert(14)
heap.insert(18)
heap.insert(19)
heap.insert(21)
heap.insert(33)
heap.insert(17)
heap.insert(27)
heap.insert(7)
heap.delmin()

print(heap.heaplist)
            
```
